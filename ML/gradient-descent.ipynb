```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(stpt,learn_rate,tol,max_iter):
  x = stpt
  x_history = [x]
  for _ in range(max_iter):
    gradient = 2 * (x+3)
    x_new = x - learn_rate * gradient
    x_history.append(x_new)

    if(abs(x_new-x)<tol):
      break
    x = x_new
  return x,x_history
```

```python
stpt = 2
learn_rate = 0.1
tol = 1e-6
max_iter = 100
```

```python
local,x_history = gradient_descent(stpt,learn_rate,tol,max_iter)
```

```python
x_vals = np.linspace(-8,2,400)
y_vals = (x_vals+3)**2
```

```python
plt.figure(figsize=(10,10))
plt.plot(x_vals, y_vals)
plt.scatter(x_history, (np.array(x_history) + 3) ** 2, color='red')
plt.grid(True)
plt.axvline(local, color='green', linestyle='--', linewidth=1)
plt.title("Graphical representation")
```

```python
print(f'The local minima is approximately at x = {local}')
```
